```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
val_file = "data/development_kit/data/val.txt"
val_labels = [x.rstrip().split()[-1] for x in open(val_file).readlines()]
val_labels = np.array([int(x) for x in val_labels])
```

```python
csv_file = "top_5_predictions_twonetworks.validation.csv"
csv_preds = [x.rstrip().split(',')[1:] for x in open(csv_file).readlines()][1:]
csv_preds = np.array([np.array(x).astype(int) for x in csv_preds])
```

```python
print csv_preds
print len(csv_preds)
print csv_preds[:,0].shape
```

```python
print val_labels
print len(val_labels)
```

```python
conf_mat = confusion_matrix(val_labels,csv_preds[:,0])
print conf_mat
```

```python
class_file = "data/development_kit/data/categories.txt"
cats = [x.rstrip().split() for x in open(class_file).readlines()]
class_dict = {}
for x in cats:
    k = x[0].split("/")[2:]
    s = ""
    for val in k:
        s += val + "_"
    v = int(x[1])
    class_dict[v]=s
print class_dict
```

```python
%matplotlib inline
plt.figure(figsize=(10,10))
#plt.xticks(np.arange(100))
#plt.yticks(np.arange(100))
plt.imshow(conf_mat)
```

```python
#100 samples per class
#rows represent true class
for i in range(len(conf_mat)):
    true_perc = float(conf_mat[i][i])/100.
    print "Class " + class_dict[i] +": " + "True Percentage: " + str(true_perc)
```

```python
confused_classes = ['museum_indoor_','river_']
for i in range(len(conf_mat)):
    if class_dict[i] == "river_":
        print "True Class: " + str(class_dict[i])
        percentages = [float(conf_mat[i][j])/100. for j in range(len(conf_mat))]
        for p in range(len(percentages)):
            print "Class: " + str(class_dict[p]) + " Percentage: " + str(percentages[p])
        print "**********"
```
